## Vraag 5 [3 pt]

Aan een bepaalde instelling voor hoger onderwijs ontvangt iedereen met een eindscore van meer dan 85 in het afstudeerjaar een premie van 1000 EUR en iedereen met een eindscore tussen 80 en 85 ontvangt 500 EUR. Alle anderen krijgen niets. De eindscores zijn normaal verdeeld met een gemiddelde van 63 en een standaardafwijking van 12.

Stel dat er 5000 studenten afstuderen, bereken dan de totale kost voor het verstrekken van de premies voor deze instelling.

Je kan dit berekenen in een drietal stappen. Vermeld duidelijk de gebruikte stappen.

Aantal studenten: 5000

Gemiddelde (μ): 63

Standaardafwijking (σ): 12

Premie 1 (€1000): score > 85

Premie 2 (€500): score tussen 80 en 85

Stap 1: Kans en aantal studenten voor premie van €1000 (> 85)

We gebruiken de stats.norm.sf functie (Survival Function) om de kans te vinden dat een score groter dan 85 is.

Stap 2: Kans en aantal studenten voor premie van €500 (tussen 80 en 85)

We berekenen het verschil tussen de cumulatieve kans tot 85 en de cumulatieve kans tot 80 met de stats.norm.cdf functie.

Stap 3: Totale kost berekenen

We vermenigvuldigen de berekende studentenaantallen met hun respectievelijke premies.


```python
import scipy.stats as stats

# Gegeven waarden
mu = 63
sigma = 12
aantal_studenten = 5000

# Stap 1: Premie 1 (> 85)
kans_premie1 = stats.norm.sf(85, loc=mu, scale=sigma)
studenten_premie1 = kans_premie1 * aantal_studenten

# Stap 2: Premie 2 (80-85)
kans_premie2 = stats.norm.cdf(85, loc=mu, scale=sigma) - stats.norm.cdf(80, loc=mu, scale=sigma)
studenten_premie2 = kans_premie2 * aantal_studenten

# Stap 3: Totale kost
kost_premie1 = studenten_premie1 * 1000
kost_premie2 = studenten_premie2 * 500
totale_kost = kost_premie1 + kost_premie2

print(f"Verwacht aantal studenten met score > 85: {studenten_premie1:.0f}")
print(f"Verwacht aantal studenten met score tussen 80 en 85: {studenten_premie2:.0f}")
print(f"De totale verwachte kost voor de premies is: €{totale_kost:.2f}")
```

### Antwoord Vraag 5

- Verwacht aantal studenten met score > 85: 167
- Verwacht aantal studenten met score tussen 80 en 85: 225
- De totale verwachte kost voor de premies is: €279166.78

```python
import numpy as np
import scipy.stats as stats

schroot = [25.70, 24.98, 24.30, 24.43, 23.21, 23.74, 24.03, 26.76, 23.83, 24.02,  25.34, 26.12, 26.46, 23.61, 24.76, 26.92, 25.05, 24.59, 24.47, 25.37, 26.28, 22.50, 24.13, 24.63, 25.08, 24.59, 20.90, 26.70, 23.07, 19.94, 24.30, 22.53, 24.09, 23.63, 23.84]

pop_mean = 25.5

t_statistic, p_value = stats.ttest_1samp(a=schroot, popmean=pop_mean, alternative='less')

print(f"De berekende t-waarde is: {t_statistic:.4f}")

print(f"De eenzijdige p-waarde is: {p_value:.8f}")
```

3. De berekende t-waarde is: -4.3039


4. De eenzijdige p-waarde is: 0.00006721


5. Besluit: De p-waarde (0.0013) is kleiner dan het significantieniveau α van 1% (0.01). Daarom 

verwerpen we de nulhypothese (H₀) .

2. Sinds 2013 ontwaren we een stijgende trend. Wat is het groeipercentage per jaar sinds 2013?

```python

n_2013 = dfzwartbont.loc[dfzwartbont['jaar'] == 2013, 'aantal'].values[0]
n_2023 = dfzwartbont.loc[dfzwartbont['jaar'] == 2023, 'aantal'].values[0]

groeifactor = (n_2023 / n_2013) ** (1 / (2023 - 2013))
groeipercentage = (groeifactor - 1) * 100

groeipercentage
```

| ------ | ------ | ------ |

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

dfmachines = pd.DataFrame(data={
    'bedrijf': ['A', 'B', 'C', 'D'],
    'totaal aantal verkochte machines': [145, 140, 120, 105],
    'percentage dat gerepareerd moet worden': [0.10, 0.05, 0.03, 0.12]
}).set_index('bedrijf')

# Bereken de absolute aantallen
dfmachines['Gerepareerd'] = np.round(dfmachines['totaal aantal verkochte machines'] * dfmachines['percentage dat gerepareerd moet worden']).astype(int)
dfmachines['Niet gerepareerd'] = dfmachines['totaal aantal verkochte machines'] - dfmachines['Gerepareerd']

# Maak de kruistabel
kruistabel = dfmachines[['Gerepareerd', 'Niet gerepareerd']]

print("Berekende Kruistabel:")
print(kruistabel)

# Voer de Chi-kwadraattoets uit
chi2_stat, p_waarde, dof, expected = chi2_contingency(kruistabel)

print(f"\nDe p-waarde is: {p_waarde:.4f}")
```
