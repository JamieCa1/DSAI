## Exercise 2.1 - Cholesterol 

A health survey between 1988 and 1994 indicated that the mean cholesterol value for women between the age of 20 and 29 was 183 mg/dl, with a standard deviation equal to 36. We now take a random sample of 81 women.

1. Plot the probability distribution of the sample mean $\overline{x}$.
2. What is the probability of the sample mean being $\overline{x} < 185$? (Answer ≈ 69,1%)
3. What is the probability of $175 < \overline{x} < 185$ (Answer ≈ 66,9%)
4. What is the probability of $\overline{x} > 190$ (Answer ≈ 4,0%)

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

mu = 183
sigma = 36
n = 81
sigma_2 = sigma / np.sqrt(n)


x = np.linspace(mu - 4*sigma_2, mu + 4*sigma_2, 100)
y = stats.norm.pdf(x, mu, sigma_2)

plt.plot(x, y)
plt.title('Kansverdeling van het steekproefgemiddelde')
plt.xlabel('Cholesterolwaarde (mg/dl)')
plt.ylabel('Kansdichtheid')
plt.show()

prob_less_than_185 = stats.norm.cdf(185, mu, sigma_2)
print(f"Kans dat het steekproefgemiddelde kleiner is dan 185: {prob_less_than_185:.2%}")

prob_between_175_and_185 = stats.norm.cdf(185, mu, sigma_2) - stats.norm.cdf(175, mu, sigma_2)
print(f"Kans dat het steekproefgemiddelde : {prob_between_175_and_185:.2%}")


prob_greater_than_190 = 1 - stats.norm.cdf(190, mu, sigma_2)
print(f"Kans dat het steekproefgemiddelde groter is dan 190: {prob_greater_than_190:.2%}")
```

## Exercise 2.2

A random sample of 64 is drawn from a population with unknown distribution. However, the expectation and standard deviation of the population are known: $\mu = 20$ and $\sigma = 16$.

1. Plot the probability distribution of the sample mean.
2. Calculate the z-score for $\overline{x_1} = 15.5$  
3. Calculate the z-score for $\overline{x_2} = 23$ 
4. Determine the probability that 16 < $\overline{x}$ < 22 (Answer ≈ 81.9%)

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

mu = 20
sigma = 16
n = 64
sigma_2 = sigma / np.sqrt(n)

x = np.linspace(mu - 4*sigma_2, mu + 4*sigma_2, 100)
y = stats.norm.pdf(x, mu, sigma_2)

plt.plot(x, y)
plt.title('Kansverdeling van het steekproefgemiddelde')
plt.xlabel('Gemiddelde waarde')
plt.ylabel('Kansdichtheid')
plt.show()

x1 = 15.5
z_score_x1 = (x1 - mu) / sigma_2
print(f"Z-score voor x̄₁ = 15.5: {z_score_x1:.2f}")

x2 = 23
z_score_x2 = (x2 - mu) / sigma_2
print(f"Z-score voor x̄₂ = 23: {z_score_x2:.2f}")

prob_between_16_and_22 = stats.norm.cdf(22, mu, sigma_2) - stats.norm.cdf(16, mu, sigma_2)
print(f"Kans dat 16 < x̄ < 22: {prob_between_16_and_22:.2%}")
```
