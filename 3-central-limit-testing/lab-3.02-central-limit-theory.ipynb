## Exercise 2.1 - Cholesterol 

A health survey between 1988 and 1994 indicated that the mean cholesterol value for women between the age of 20 and 29 was 183 mg/dl, with a standard deviation equal to 36. We now take a random sample of 81 women.

1. Plot the probability distribution of the sample mean $\overline{x}$.
    ` bij "Plot the probability distribution... " gebruiken wij altijd PDF functie`
2. What is the probability of the sample mean being $\overline{x} < 185$? (Answer ≈ 69,1%)
3. What is the probability of $175 < \overline{x} < 185$ (Answer ≈ 66,9%)
4. What is the probability of $\overline{x} > 190$ (Answer ≈ 4,0%)

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Parameters
mu = 183           # Populatiegemiddelde
sigma = 36         # Populatiestandaardafwijking
n = 81             # Steekproefgrootte

mu_sample = mu
sigma_sample = sigma / np.sqrt(n)  # Standaardfout = 4

# 1. Plot de kansverdeling van het steekproefgemiddelde
x = np.linspace(mu_sample - 4*sigma_sample, mu_sample + 4*sigma_sample, 1000)
pdf = stats.norm.pdf(x, mu_sample, sigma_sample)

plt.figure(figsize=(8,5))
plt.plot(x, pdf, label="Verdeling van steekproefgemiddelde")
plt.title("Verdeling van het steekproefgemiddelde (x̄)")
plt.xlabel("Cholesterolwaarde (mg/dl)")
plt.ylabel("Kansdichtheid")
plt.grid()
plt.legend()
plt.show()

# 2. P(x̄ < 185)
p_2 = stats.norm.cdf(185, mu_sample, sigma_sample)
print(f"2. P(x̄ < 185) = {p_2:.4f} → {p_2*100:.1f}%")

# 3. P(175 < x̄ < 185)
p_3 = stats.norm.cdf(185, mu_sample, sigma_sample) - stats.norm.cdf(175, mu_sample, sigma_sample)
print(f"3. P(175 < x̄ < 185) = {p_3:.4f} → {p_3*100:.1f}%")

# 4. P(x̄ > 190)
p_4 = 1 - stats.norm.cdf(190, mu_sample, sigma_sample)
print(f"4. P(x̄ > 190) = {p_4:.4f} → {p_4*100:.1f}%")
```

## Exercise 2.2

A random sample of 64 is drawn from a population with unknown distribution. However, the expectation and standard deviation of the population are known: $\mu = 20$ and $\sigma = 16$.

1. Plot the probability distribution of the sample mean.
2. Calculate the z-score for $\overline{x_1} = 15.5$  
3. Calculate the z-score for $\overline{x_2} = 23$ 
4. Determine the probability that 16 < $\overline{x}$ < 22 (Answer ≈ 81.9%)

 - `De z-score wordt berekend met de formule:`

$$
z = \frac{\overline{x} - \mu}{\frac{\sigma}{\sqrt{n}}}
$$


```python
import scipy.stats as stats
import matplotlib.pyplot as plt

sample = 64
mu = 20
sigma = 16
mu_sample = mu
sigma_sample = sigma/ np.sqrt(sample)

x = np.linspace(mu_sample - 4 * sigma_sample, mu_sample + 4 * sigma_sample, 1000)
pdf = stats.norm.pdf(x,mu_sample,sigma_sample)

plt.plot(x, pdf, label="PDF van x̄")
plt.title("Verdeling van het steekproefgemiddelde x̄")
plt.xlabel("x̄")
plt.ylabel("Kansdichtheid")
plt.grid()
plt.legend()
plt.show()


# 2. Z-score voor x̄ = 15.5
x1 = 15.5
z1 = (x1 - mu) / sigma_sample
print(f"Z-score voor x̄ = 15.5 is {z1:.2f}")


# 3. Z-score voor x̄ = 23
x2 = 23
z2 = (x2 - mu) / sigma_sample
print(f"Z-score voor x̄ = 23 is {z2:.2f}")


p_lower = stats.norm.cdf(22, mu_sample, sigma_sample)
p_upper = stats.norm.cdf(16, mu_sample, sigma_sample)
prob = p_lower - p_upper
print(f"P(16 < x̄ < 22) =  {prob*100:.1f}%")
```
